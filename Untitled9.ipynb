```python
import os
import csv
import datetime
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import RandomForestClassifier

data = []
tag3_up = []
tag1_up = []

for t in ['pnj.txt', 'sab.txt', 'ssi.txt', 'cii.txt', 'ctd.txt',
         'nvl.txt', 'vnm.txt', 'gmd.txt', 'sbt.txt', 'ctg.txt',
         'vic.txt', 'ree.txt', 'vjc.txt', 'vre.txt', 'dpm.txt',
         'hpg.txt', 'msn.txt', 'mwg.txt', 'hdb.txt', 'mbb.txt',
         'ros.txt', 'vpb.txt', 'gas.txt', 'stb.txt', 'vhm.txt',
         'dhg.txt', 'tcb.txt', 'fpt.txt', 'vcb.txt', 'eib.txt']:
    cpdata = []
    with open('../data/VN30/' + t) as csvfile:
        reader = csv.DictReader(csvfile)
        for row in reader:
            cp = []
            cp.append(float(row['<Open>']))
            cp.append(float(row['<High>']))
            cp.append(float(row['<Low>']))
            cp.append(float(row['<Close>']))
            cp.append(float(row['<Volume>']))
            cpdata.append(cp)

    cpdata.reverse()

    n = 9
    print(t, "...")
    for i in range(0, len(cpdata) - (n + 3)):
        prodata = cpdata[i : i + (n + 4)]
        subdata = []
        for j in range(0, n):
            subdata.extend(prodata[j][:-1])
        subsubdata = []
        for j in range(0, n * 4):
            for k in range(j + 1, n * 4):
                if(subdata[k] > subdata[j]):
                    subsubdata.append(1)
                elif(subdata[k] == subdata[j]):
                    subsubdata.append(0)
                else:
                    subsubdata.append(-1)
        for j in range(0, n):
            for k in range(j + 1, n):
                if(prodata[k][4] > prodata[j][4]):
                    subsubdata.append(1)
                elif(prodata[k][4] == prodata[j][4]):
                    subsubdata.append(0)
                else:
                    subsubdata.append(-1)
        data.append(subsubdata)
        if(prodata[n + 3][0] > prodata[n][3]):
            tag3_up.append(1)
        else:
            tag3_up.append(0)
        if(prodata[n + 1][0] > prodata[n][0]):
            tag1_up.append(1)
        else:
            tag1_up.append(0)
```

```python
clf1 = []
rs = 0
while(rs < 1000):
    clf = RandomForestClassifier(random_state=rs)
    clf.fit(data, tag1_up)
    clf1.append(clf)
    rs += 1
```
