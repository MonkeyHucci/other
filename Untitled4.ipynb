```python
import os
import csv
import datetime
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import RandomForestClassifier

data = []
tag0_low = []
tag3_up = []

for t in ['pnj.txt', 'sab.txt', 'ssi.txt', 'cii.txt', 'ctd.txt',
         'nvl.txt', 'vnm.txt', 'gmd.txt', 'sbt.txt', 'ctg.txt',
         'vic.txt', 'ree.txt', 'vjc.txt', 'vre.txt', 'dpm.txt',
         'hpg.txt', 'msn.txt', 'mwg.txt', 'hdb.txt', 'mbb.txt',
         'ros.txt', 'vpb.txt', 'gas.txt', 'stb.txt', 'vhm.txt',
         'dhg.txt', 'tcb.txt', 'fpt.txt', 'vcb.txt', 'eib.txt',
         'yeg.txt']:
    cpdata = []
    with open('../data/VN30/' + t) as csvfile:
        reader = csv.DictReader(csvfile)
        for row in reader:
            cp = []
            cp.append(float(row['<Open>']))
            cp.append(float(row['<High>']))
            cp.append(float(row['<Low>']))
            cp.append(float(row['<Close>']))
            cpdata.append(cp)

    cpdata.reverse()

    n = 10
    print(t, "...")
    for i in range(0, len(cpdata) - (n + 3)):
        prodata = cpdata[i : i + (n + 4)]
        subdata = []
        for j in range(1, n):
            subdata.extend(round(x / prodata[j - 1][3], 3) for x in prodata[j][0:])
        data.append(subdata)
        tag0_low.append(round(prodata[n][2] / prodata[n - 1][3], 3))
        if(prodata[n + 3][1] > prodata[n][2]):
            tag3_up.append(1)
        else:
            tag3_up.append(0)
```

```python
clf1 = []
rs = 0
while(rs < 100):
    clf = RandomForestRegressor(random_state=rs)
    clf.fit(data, tag0_low)
    clf1.append(clf)
    rs += 1

clf2 = []
rs = 0
while(rs < 100):
    clf = RandomForestClassifier(random_state=rs)
    clf.fit(data, tag3_up)
    clf2.append(clf)
    rs += 1
```
